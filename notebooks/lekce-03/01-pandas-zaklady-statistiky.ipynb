## Metoda .describe()

Metoda `describe()` v Pandas poskytuje rychlý přehled o základních deskriptivních statistikách pro daný DataFrame. 

Tato metoda je užitečná pro získání rychlého pochopení rozložení dat, zejména u numerických sloupců. 

Typické míry střední hodnoty jsou: 
- aritmetický průměr – průměr všech hodnot ve vzorku (počítáno podle hodnot), 
- medián – hodnota, které se nachází ve středu vzorku (počítáno podle pozic), a 
- modus – nejčastěji se vyskytující hodnota (počítáno podle frekvencí).

Typické míry rozptylu jsou: 
- kvartily – každá ze tří hodnot, které rozdělují tříděné údaje na čtyři stejné části tak, aby každá část představovala jednu čtvrtinu zkoumaného vzorku: 

    - První kvartil (neboli dolní kvartil) odděluje nejnižších 25 % dat (25. percentil), 
    - druhý kvartil (neboli střední) odděluje data v polovině (50. percentil), a 
    - třetí kvartil (neboli horní kvartil) odděluje nejvyšších 25 % dat (75. percentil).

Typické míry extrémů jsou: 
- minimum – nejmenší číslo vzorku, a 
- maximum – největší číslo vzorku.

```python
# Základní deskriptivní statistiky
print("Základní statistické ukazatele:")
data.describe()
```

## Směrodatná odchylka a rozptyl

Rozptyl a smerodatná odchylka se používají k popisu rozložení dat a míry variability v těchto datech. 

```python
column_name = "Calories"
```

Vztah mezi nimi je následující:

- Rozptyl (Variance): Rozptyl je míra variability nebo rozptýlení hodnot v datasetu. Je to střední hodnota kvadrátů odchylek od průměru. Matematicky je rozptyl definován jako průměr čtverců rozdílů mezi každou hodnotou a průměrem:

- Smerodatná odchylka (Standard Deviation): Smerodatná odchylka je **odmocnina z rozptylu**. Je to míra variability v datech, ale v původních jednotkách. Smerodatná odchylka ukazuje, jak daleko jsou hodnoty průměrné hodnoty rozloženy. 

Takže smerodatná odchylka je rozptyl v "rozumných" jednotkách, což z něj činí více interpretabilní míru variability. 


```python
print("Rozptyl:", data[column_name].var())
print("Směrodatná odchylka:", data[column_name].std())
```

```python
# Krok 1: Vypočtěte průměr
prumer = np.mean(data[column_name])

# Krok 2: Vypočtěte rozptyl
rozptyl = np.sum((data[column_name] - prumer) ** 2) / (len(data[column_name]) - 1)
# rozptyl = np.mean((data[column_name] - prumer) ** 2) # bez Besselovy korekce

# Krok 3: Vypočtěte směrodatnou odchylku
smerodatna_odchylka = np.sqrt(rozptyl)

print("Rozptyl:", rozptyl)
print("Směrodatná odchylka:", smerodatna_odchylka)

print(
    f"\nHodnoty v datové sadě se v průměru liší o {round(smerodatna_odchylka)} jednotek od jejich střední hodnoty."
)
```

**Míra rozptylu:** Vyšší hodnota SD ukazuje větší rozptyl dat - tj. data jsou více rozptýlená od průměru. Naopak, nižší hodnota SD naznačuje, že data jsou blíže k průměrné hodnotě.

**Normalita dat:** Pokud jsou data normálně rozdělená (gaussovská distribuce), pak asi 68% hodnot leží do jedné směrodatné odchylky od průměru, 95% hodnot leží do dvou směrodatných odchylek a 99,7% hodnot leží do tří směrodatných odchylek.

**Porovnání datových sad:** Směrodatná odchylka umožňuje porovnání rozptylu dvou a více datových sad. Pokud má jedna sada SD 240 a druhá má SD 10, první sada je mnohem více rozptýlená.


```python
column_name = "Calories"
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Výpočet směrodatné odchylky
std_dev = data[column_name].std()  # Vypočítá směrodatnou odchylku
mean_value = np.mean(data[column_name])  # Vypočítá průměrnou hodnotu

# Vykreslení histogramu a přizpůsobení normálního rozdělení pro skutečná data
plt.figure(figsize=(10, 6))  # Nastavení velikosti obrázku

# Histogram pro skutečná data
sns.histplot(
    data[column_name],
    color="red",  # Barva histogramu
    kde=False,  # Nezobrazuje odhad hustoty pravděpodobnosti (Kernel Density Estimate)
    stat="density",  # Normalizace tak, aby plocha pod grafem odpovídala hustotě pravděpodobnosti
    bins=20,  # Počet sloupců v histogramu
    label=f"SD = {std_dev:.2f}",  # Popisek zobrazující směrodatnou odchylku
)

# Vykreslení křivky normálního rozdělení pro skutečná data
# Křivka přidaná do grafu, která znázorňuje, jak by vypadala normální distribuce s daným průměrem a směrodatnou odchylkou, a slouží k porovnání s aktuálním rozložením dat
xmin, xmax = plt.xlim()  # Získání minimální a maximální hodnoty x-ové osy
x = np.linspace(xmin, xmax, 100)  # Vytvoření 100 bodů mezi xmin a xmax
p = norm.pdf(x, mean_value, std_dev)  # Vypočet hodnot normálního rozdělení pro x
plt.plot(
    x, p, "k", linewidth=2, label="Normal Dist Fit (Actual Data)"
)  # Vykreslení křivky

# Nastavení titulku, legendy a popisků os
plt.title(f"Histogram and Normal Distribution Fit for Actual {column_name} Data")
plt.legend()  # Zobrazí legendu
plt.xlabel(f"{column_name}")  # Popisek x-ové osy
plt.ylabel("Density")  # Popisek y-ové osy
plt.grid(True)  # Zapne mřížku pro lepší orientaci
plt.show()  # Zobrazí graf
```

## Kvartily a boxplot

Boxplot je grafické znázornění rozdělení dat.

Zobrazuje medián, kvartily, rozptyl a extrémní hodnoty pomocí obdélníku (boxu) a "vousů".

Více o boxplotu: <https://www.simplypsychology.org/boxplots.html>


![Box plot explained](../../img/box-whisker-plot.jpg)

Mezikvartilové rozpětí (IQR) je rozdíl mezi třetím (horním) a prvním (dolním) kvartilem datové sady.

IQR poskytuje míru statistické variability ve střední polovině dat (jak široce se hodnoty rozprostírají kolem střední hodnoty).

```python
# Výpočet kvartilů a IQR (mezíkvartilového rozpětí)
Q1 = data["Calories"].quantile(0.25)
Q3 = data["Calories"].quantile(0.75)
IQR = Q3 - Q1

print("Kvartily a IQR:")
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

```python
# Vytvoření boxplotu
plt.boxplot(data[column_name], vert=False)
plt.title(f'Boxplot sloupce "{column_name}"')
plt.xlabel("Hodnota")
plt.show()
```

## Odlehlá pozorování *aka outliers*

Odlehlá pozorování jsou data, která se významně liší od ostatních hodnot v datové sadě, často vzdálená od mediánu nebo průměru.

Odstranění odlehlých hodnot z dat může být žádoucí, protože mohou zkreslit statistické analýzy a modely, ovlivňujíce například výpočet průměru, směrodatné odchylky nebo výsledky regresních analýz, a tím potenciálně vedoucí k nesprávným závěrům.


Metoda mezíkvartilového rozpětí (IQR) identifikuje a odstraňuje odlehlé hodnoty tak, že určí hranice jako 1.5krát IQR nad horním kvartilem a pod dolním kvartilem a považuje hodnoty mimo tyto hranice za odlehlé.

```python
# Definice mezí pro outliery
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Vyfiltrování outlierů
filtered_data = data[
    (data[column_name] >= lower_bound) & (data[column_name] <= upper_bound)
]

print(f"Původní DataFrame {data.shape}")
print(f"DataFrame bez outlierů: {filtered_data.shape}")
```

```python
# Vytvoření boxplotu bez outliers
plt.boxplot(filtered_data[column_name], vert=False)
plt.title(f'Boxplot sloupce "{column_name}"')
plt.xlabel("Hodnota")
plt.show()
```
