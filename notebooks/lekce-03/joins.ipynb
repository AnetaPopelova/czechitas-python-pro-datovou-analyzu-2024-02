```python
import pandas as pd
```

```python
# Vytvoření DataFrame 'mazlicci'
mazlicci = pd.DataFrame({
    'mazlicek_id': [1, 2, 3, 4],
    'jmeno': ['Rex', 'Mia', 'Oreo', 'Bella'],
    'typ': ['Pes', 'Kočka', 'Pes', 'Kočka']
})

# Vytvoření DataFrame 'hracky'
hracky = pd.DataFrame({
    'hracka_id': ['H1', 'H2', 'H3', 'H4'],
    'nazev_hracky': ['Kost', 'Pískací myš', 'Lanové hračky', 'Míček'],
    'mazlicek_id': [1, 2, 1, 4]
})
```

```python
# Inner Join
inner_join_mazlicci = pd.merge(mazlicci, hracky, on='mazlicek_id', how='inner')
inner_join_mazlicci
```

```python
# Left Join
left_join_mazlicci = pd.merge(mazlicci, hracky, on='mazlicek_id', how='left')
left_join_mazlicci
```

```python
# Right Join
right_join_mazlicci = pd.merge(mazlicci, hracky, on='mazlicek_id', how='right')
right_join_mazlicci
```

```python
# Outer Join
outer_join_mazlicci = pd.merge(mazlicci, hracky, on='mazlicek_id', how='outer')
outer_join_mazlicci
```
